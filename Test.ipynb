```python
import numpy as np
import pandas as pd
import pickle
import nltk

from tensorflow.keras.models import load_model
from nltk.corpus import stopwords
```

```python
LABELS = ['business', 'sport', 'tech', 'politics', 'entertainment']
```

```python
model = load_model('./domainClassifier.h5')
```

```python
vectorizer = pickle.load(open('./vectorizer.pickle', 'rb'))
```

```python
import os
import glob
```

```python
current_dir = os.getcwd()
```

```python
def find_label(pred):
    for i, element in enumerate(pred):
        for j, item in enumerate(element):
            if pred[i][j]==np.max(element):
                return LABELS[j]
```

```python
path = ''
path = current_dir + "\\bbc\\tech"
os.chdir(path)
```

```python
out = []
```

```python
for files in glob.glob("*.txt"):
    with open(files, 'r') as f:
        text = ''
        data = []
        lines = [l.strip() for l in f.readlines()]
        for line in lines:
            if len(line)>0:
                text += line
        if(len(text)>0):
            data.append(text)
            data = vectorizer.transform(data)
            pred = model.predict(data)
            label = find_label(pred)
            out.append(label)
```
